```python
import pandas
```

```python
zam_praha = pandas.read_csv("zam_praha.txt")
zam_plzen = pandas.read_csv("zam_plzeň.txt")
zam_liberec = pandas.read_csv("zam_liberec.txt")

zam_praha["mesto"] = "Praha"
zam_plzen["mesto"] = "Plzeň"
zam_liberec["mesto"] = "Liberec"

zamestnanci = pandas.concat([zam_praha, zam_plzen, zam_liberec], ignore_index=True)
zamestnanci.to_csv("zamestnanci.csv", index=False)
zamestnanci.head()
```

```python
platy_2021_02 = pandas.read_csv("platy_2021_02.txt")
platy_2021_02.head()
```

```python
zam_platy = pandas.merge(zamestnanci, platy_2021_02, on=["cislo_zamestnance"])
zam_platy
```

```python
print(zam_platy.groupby("mesto")["plat"].mean())
```
